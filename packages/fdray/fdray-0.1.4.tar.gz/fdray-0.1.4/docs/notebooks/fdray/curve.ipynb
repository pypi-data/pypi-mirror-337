```python
import numpy as np

from fdray import *


def render(*attrs):
    return Scene(
        Camera(0, 20, view_scale=1.1, look_at=(0, 0.5, 0.5)),
        LightSource(0, "white"),
        *attrs,
    ).render(300, 300)
```

```python
a = np.array([[0, 0, 0], [0, 0, 1], [0, 1, 1], [0, 1, 0]])
x = Polyline(a, 0.1, Color("cyan"))
render(x)
```

```python
x = Curve(a, 0.1, Color("cyan"))
render(x)
```
