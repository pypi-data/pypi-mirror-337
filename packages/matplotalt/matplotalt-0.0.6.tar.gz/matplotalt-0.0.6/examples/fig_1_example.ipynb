```python
import matplotlib.pyplot as plt
from matplotalt import show_with_alt

sunshine_hours = [69, 108, 178, 207, 253, 268, 312, 281, 221, 142, 72, 52]
months = ["Jan", "Feb", "Mar", "Apr", "May", "June", \
          "July", "Aug", "Sep", "Oct", "Nov", "Dec"]
plt.bar(months, sunshine_hours)
plt.title("Average yearly hours of sunshine")
plt.xlabel("Month")
plt.ylabel("Hours of sunshine")
plt.gcf().set_size_inches(6, 3)


# Generate and surface heuristic alt text
show_with_alt(desc_level=3, methods=["markdown", "img_file"])
```
