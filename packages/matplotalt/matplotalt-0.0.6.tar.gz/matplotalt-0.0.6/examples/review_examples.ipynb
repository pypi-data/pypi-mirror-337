# Heatmap

```python
# imports
import matplotlib.pyplot as plt
import numpy as np
from matplotalt import show_with_alt, generate_alt_text, surface_alt_text


# Data for the heatmap
months = ["Jan", "Feb", "Mar", "Apr", "May", "June", "July", "Aug", "Sep", "Oct", "Nov", "Dec"]
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
np.random.seed(42)
data = np.random.randint(0, 100, size=(len(days), len(months)))


# Create a heatmap
fig, ax = plt.subplots(figsize=(10, 6))
cax = ax.imshow(data, cmap="YlGnBu", aspect="auto")


# Add colorbar
cbar = fig.colorbar(cax, ax=ax)
cbar.set_label("Values", fontsize=12)


# Set axis labels and ticks
ax.set_xticks(np.arange(len(months)))
ax.set_yticks(np.arange(len(days)))
ax.set_xticklabels(months)
ax.set_yticklabels(days)
ax.set_xlabel("Months", fontsize=12)
ax.set_ylabel("Days", fontsize=12)
ax.set_title("Heatmap Example with Random Data", fontsize=14)


# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)
plt.tight_layout()


# Show the heatmap
# Comment out plt.show()
#plt.show()
show_with_alt(methods="markdown")
```

# Bar

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotalt import show_with_alt, generate_alt_text, surface_alt_text


# Generate fake data
np.random.seed(42) # For reproducibility
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
values = [np.random.randint(50, 100) for _ in categories]


# Create a bar chart
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='lightblue', edgecolor='blue')


# Add labels and title
plt.title("Bar Chart Example with Fake Data", fontsize=14)
plt.xlabel("Categories", fontsize=12)
plt.ylabel("Values", fontsize=12)


# Show the plot
plt.tight_layout()
# Comment out plt.show()
#plt.show()

show_with_alt(methods="markdown")
```
